```python
import numpy as np
data = ["16,1,2,0,4,2,7,1,2,14"]
with open('input.txt') as f:
    data = f.read().splitlines()
new_data =[]
for element in data:
  new_data.append(element.split(","))
```

```python
data = []
for lista in new_data:
  for element in lista:
    element = int(element)
    data.append(element)
```

```python
data = np.array(data)
data
```

```python
data_min = min(data)
data_max = max(data)
```

```python
elements = range(data_min,data_max+1,1)
```

```python
sum_tabs = 0 
# for i in tab_with_all_step:
#   sum_tabs += i.sum()
```

```python
range_value = []
tabs = []
for idx,value in enumerate(elements):
  #sum_tab.append(abs(data - value).sum())
  new_value = abs(value - data)
  tabs =[]
  for element_value in new_value:
    tabs.append(element_value)
  range_value.append(tabs)

```

```python
sum_elements = 0
list_values = []
for idx,value in enumerate(range_value):
  sum_elements = 0
  for v in value:
    el = sum(list(range(v+1)))
    sum_elements += el
  list_values.append(sum_elements)
  
```
