```python
import pandas as pd
test = ["00100", "11110", "10110", "10111", "10101", "01111", "00111", "11100", "10000", "11001", "00010", "01010"]
with open('input.txt') as f:
    test = f.read().splitlines()
test_as_values = []
len_test = len(test)

for element in test:
  test_as_values.append((list(element)))
df = pd.DataFrame(test_as_values).astype(int)
df_copy_oxygen  = df.copy()
```

```python
rows = len(df_copy_oxygen) 
cols =  len(df_copy_oxygen.columns)
i = 0 
while i!=cols: 
  check = int(len(df_copy_oxygen))
  mask = (df_copy_oxygen.sum()/check>=0.5)
  print(mask,check)
  if mask[i] == True: 
    df_copy_oxygen= df_copy_oxygen.drop(df_copy_oxygen.loc[df_copy_oxygen[i]==0].index, inplace=False)
  elif mask[i] == False:
    df_copy_oxygen = df_copy_oxygen.drop(df_copy_oxygen.loc[df_copy_oxygen[i]==1].index, inplace=False)
  i+=1
  rows = len(df_copy_oxygen) 
```

```python
df_copy_co2  = df.copy()
rows = len(df_copy_co2) 
cols =  len(df_copy_co2.columns)
i = 0 
while i!=cols: 
  check = int(len(df_copy_co2))
  mask = (df_copy_co2.sum()/check>=0.5)
  print(mask,check)
  if len(df_copy_co2) == 1:
    break
  if mask[i] == True: 
    df_copy_co2= df_copy_co2.drop(df_copy_co2.loc[df_copy_co2[i]==1].index, inplace=False)
  elif mask[i] == False:
    df_copy_co2 = df_copy_co2.drop(df_copy_co2.loc[df_copy_co2[i]==0].index, inplace=False)
  i+=1
  rows = len(df_copy_co2) 
```

```python
oxygen = df_copy_oxygen.values
co2 = df_copy_co2.values
binary_to_decimal = [2**x for x in range(len(oxygen[0])-1,-1, -1)]
```

```python
decimal_oxygen = []
decimal_co2 = []
for idx,binary in enumerate(binary_to_decimal):
  values_oxygen= binary*oxygen[0][idx]
  decimal_oxygen.append(values_oxygen)
  values_co2 = binary*co2[0][idx]
  decimal_co2.append(values_co2)
```

```python
sum(decimal_co2) * sum(decimal_oxygen)
```
